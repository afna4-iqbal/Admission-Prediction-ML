```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
data=pd.read_csv('/content/Admission_Predict.csv')
data
```

```python
data.isna().sum()
```

```python
data.columns
```

```python
data.dtypes
```

```python
data.drop('Serial No.',axis=1,inplace=True) #axis=1(vertically dropped) #axis=0(horizontally drpped)
data
```

```python
x=data.drop('Chance of Admit ', axis=1)
y=data['Chance of Admit ']
x
y
```

```python
data.shape
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23)#random state is parameter it makes training and testing data become unique
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python

model=LinearRegression()
model.fit(x_train,y_train) #training models
```

```python
y_pred=model.predict(x_test)
```

```python
pd.DataFrame({'True value':y_test,'Predict value':y_pred})
```

```python
#To find Errors
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred) #check the accuray (o/p become 0 or 1) if o/p shows 1 it is 95% accuracy
```
